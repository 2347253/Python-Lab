```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

import re


root = tk.Tk()
root.title("Domain Details Form")
root.geometry("520x530")

def submit_form():
    # Retrieve values from input fields
    id_value = id_entry.get()
    name_value = name_entry.get()
    age_value = gender_var.get()
    is_active_value = is_active_var.get()
    game_value= game_var.get()
    phone_value=phone_var.get()

    if not validate_email(game_value):
        messagebox.showerror("Error", "Invalid Email Address")
        return
    if not validate_phone(phone_value):
        messagebox.showerror("Error", "Invalid Phone Number (up to 10 digits)")
        return
    if not validate_username(name_value):
        messagebox.showerror("Error", "Invalid Username (letters only)")
        return
    
    tree.insert("", "end", values=(id_value, name_value, game_value, phone_value, age_value, is_active_value))  #game value is Email Value and
    # age value is gender value 
    # Clear the input fields
    id_entry.delete(0, tk.END)
    name_entry.delete(0, tk.END)
    phone_var.delete(0,tk.END)
    game_var.delete(0,tk.END)
    gender_combo.set("")  # Clear the age selection
    is_active_var.set("")

def validate_id(value):
  """Returns True if the value is a valid ID card number, False otherwise."""
  try:
    int(value)
    return True
  except ValueError:
    return False
  
def validate_email(email):
    """Returns True if the email is valid, False otherwise."""
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return re.match(pattern, email) is not None

def validate_phone(phone):
    """Returns True if the phone number is valid (up to 10 digits), False otherwise."""
    pattern = r'^\d{10}$'
    return re.match(pattern, phone) is not None

def validate_username(username):
    """Returns True if the username contains only letters, False otherwise."""
    pattern = r'^[a-zA-Z]+$'
    return re.match(pattern, username) is not None


frame = ttk.Frame(root, padding=20)                            
frame.grid(column=0, row=0, sticky=(tk.W, tk.E, tk.N, tk.S))

id_label = ttk.Label(frame, text="PLayer ID:")
id_label.grid(column=0, row=0, sticky=tk.W)
id_entry = ttk.Entry(frame,validate='key',validatecommand=(frame.register(validate_id),'%S'))
id_entry.grid(column=1, row=0, padx=5, pady=5, sticky=tk.W)

name_label = ttk.Label(frame, text="User Name:")
name_label.grid(column=0, row=1, sticky=tk.W)
name_entry = ttk.Entry(frame)
name_entry.grid(column=1, row=1, padx=5, pady=5, sticky=tk.W)

game_label = ttk.Label(frame, text="Email ID:")
game_label.grid(column=0, row=2, sticky=tk.W)
game_var = ttk.Entry(frame)
game_var.grid(column=1, row=2, padx=5, pady=5, sticky=tk.W)

gender_label = ttk.Label(frame, text="Gender:")
gender_label.grid(column=0, row=3, sticky=tk.W)
gender_var = tk.StringVar()
gender_combo = ttk.Combobox(frame, textvariable=gender_var, values=["Male", "Female", "Other"])
gender_combo.grid(column=1, row=3, padx=5, pady=5, sticky=tk.W)

is_active_label = ttk.Label(frame, text="Is Active:")
is_active_label.grid(column=0, row=4, sticky=tk.W)
is_active_var = tk.StringVar()
is_active_yes = ttk.Radiobutton(frame, text="Yes", variable=is_active_var, value="Yes")
is_active_yes.grid(column=1, row=4, padx=10, pady=1, sticky=tk.W)
is_active_no = ttk.Radiobutton(frame, text="No", variable=is_active_var, value="No")
is_active_no.grid(column=2, row=4, padx=0, pady=1, sticky=tk.W)

phone_label=ttk.Label(frame, text="Phone no:")
phone_label.grid(column=0, row=5, sticky=tk.W)
phone_var=ttk.Entry(frame)
phone_var.grid(column=1, row=5,sticky=tk.W)

submit_button = ttk.Button(frame, text="Submit", command=submit_form)
submit_button.grid(column=0, row=6, columnspan=1, pady=10)

table_frame = ttk.Frame(root)
table_frame.grid(column=0, row=1, sticky=(tk.W, tk.E, tk.N, tk.S))

result_label = ttk.Label(frame, text="", wraplength=300)
result_label.grid(column=0, row=5, columnspan=3, pady=10, sticky=tk.W)

tree = ttk.Treeview(table_frame, columns=("ID", "UserName","Email", "Phone_no","Gender","Is Active"),show="headings",)
tree.grid(column=0, row=0, padx=5, pady=10)

tree.heading("ID", text="ID")
tree.heading("UserName", text="UserName")
tree.heading("Gender", text="Gender")
tree.heading("Email", text="Email")
tree.heading("Phone_no", text="Phone No")
tree.heading("Is Active", text="Is Active")

tree.column("ID", width=10)
tree.column("UserName", width=100)
tree.column("Gender", width=100)
tree.column("Email", width=100)
tree.column("Phone_no", width=100)
tree.column("Is Active", width=100)

root.mainloop()
```
