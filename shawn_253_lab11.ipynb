```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ESport_Earnings.csv')
x = df['GameName']
y = df['PlayerNo'] 
plt.figure(figsize=(13, 10))  # Adjust the figure size as needed
plt.plot(x, y)
plt.title('Games and their Players')
plt.xlabel('Game Name')
plt.ylabel('Number of Players')
plt.show()

```

```python
categories = df["Top_Country"]
values = df["Top_Country_Earnings"]

# Create a bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(categories, values, color='skyblue')

plt.xlabel("Top_Country")
plt.ylabel("Top_Country_Earnings")
plt.title("Top Country earnings on Games")


plt.show()
```

```python
categories = df["GameName"]
values = df["TotalMoney"]

# Create a pie chart
plt.figure(figsize=(10, 10))  # Adjust the figure size as needed
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, )

# Add a title
plt.title("Total Money Earned by The online Games")

plt.show()
```

```python
x_values = df["GameName"]
y_values = df["PlayerNo"]

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(x_values, y_values, marker='o', c='blue', label='Data Points')

# Add labels and a title
plt.xlabel("Game Name")
plt.ylabel("Player Numbers")
plt.title("Scatter Plotting  of the Count of players based on games")

plt.show()
```
